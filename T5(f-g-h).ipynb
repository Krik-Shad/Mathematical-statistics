5. Случайная величина распределена равномерно на отрезке $[θ, 2θ]$.

a) По выборке объема $n$ найти оценки параметра $θ$ методом моментов и методом максимального правдоподобия.

b) Проверить оценки на несмещенность и состоятельность. Исправить эти оценки, если необходимо.

c) Сравнить эффективность исправленных оценок.

d) Построить точный доверительный интервал для параметра $θ$.

e) Построить асимптотический доверительный интервал для параметра $θ$.

f) Сгенерируйте выборку объема $n = 100$ для некоторого значения параметра $θ$. Вычислите указанные выше доверительные интервалы для доверительной вероятности 0.95.

g) Численно постройте бутстраповский доверительный интервал.

h) Сравнить все интервалы.

```python
#подключение всех необходимых библиотек

import numpy as np
from scipy import stats
```

```python
#Генерация выборки

def generate_sample(theta):
    y = np.random.random()
    x = theta*(1+y)
    return x
n = 100
theta = 50
sample = np.array([generate_sample(theta) for _ in range(n)])
```

```python
def calc_tochniy(sample, alpha=0.05):
    t1 = 1 + (alpha/2)**(1/n)
    t2 = 1 + (1 - alpha/2)**(1/n)
    xmax = np.max(sample)
    left = xmax/(1 + t2)
    right = xmax/(1 + t1)
    return left, right, right - left
```

```python
l1, r1, leng1 = calc_tochniy(sample)

print(f"For the precise interval:\nRight border: {r1}, left border: {l1}, the length of the interval: {leng1}")
```

```python
def calc_asympt(sample, alpha=0.05):
    z = stats.norm.ppf(1 - alpha/2)
    x_mean = np.mean(sample)
    theta_wave = 2/3 * x_mean
    left = theta_wave - z * theta_wave**2 / (27 * n)
    right = theta_wave + z * theta_wave**2 / (27 * n)
    return left, right, right-left
```

```python
l2, r2, leng2 = calc_asympt(sample)

print(f"For the asymptotic interval:\nRight border: {r2}, left border: {l2}, the length of the interval: {leng2}")
```

```python
#Бутстрап

def calc_bootstrap(sample, alpha=0.05):
    bsp_iteration = 1000
    theta_OMP = (n+1) * np.max(sample)/(2*n + 1)
    bsp_delta = np.array([(n+1)*np.max(np.random.choice(sample, size=n))/(2*n+1) - theta_OMP for _ in range(bsp_iteration)])
    bsp_sorted = np.sort(bsp_delta)
    left = theta_OMP - bsp_sorted[24]
    right = theta_OMP - bsp_sorted[974]
    return left, right, right-left
    
```

```python
l_b, r_b, leng_b = calc_bootstrap(sample)

print(f"For the Bootstrap's interval:\nRight border: {r_b}, left border: {l_b}, the length of the interval: {leng_b}")
```

```python
#Сравниваем
intervals = {"precise":leng1, "asymptotic":leng2, "bootstrap":leng_b}
intervals = sorted(intervals.items(), key=lambda item: item[1])
for interval_name, interval_value in intervals:
    print(f"{interval_name} (l = {interval_value:.2f})\n")
```
