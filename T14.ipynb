14. Пусть $x_n$ и $y_m$ — независимые случайные выборки из нормального распределения с параметрами $a, \sigma_x^2 = 2$ и $b, \sigma_y^2 = 1$ соответственно.  
Используя реализацию случайных выборок: $x = \{-1.11, -6.10, 2.42\}$,  
$y = \{-2.29, -2.91\}$, проверить гипотезу о равенстве средних против альтернатив $a > b$. Построить график мощности критерия.

```python
#Подключаем нужные библиотеки
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
alpha = 0.05 #т.к. не задано иного

n = 3
m = 2

x_n = np.array([-1.11, -6.10, 2.42])
y_m = np.array([-2.29, -2.91])

sigma_x = 2
sigma_y = 1
```

```python
delta = x_n.mean() - y_m.mean()

print(f"{delta = }")

dist_H0 = norm(
    loc=0,
    scale=np.sqrt(sigma_x / n + sigma_y / m)
    )

quantil = dist_H0.ppf(1 - alpha)

print(f"Квантиль порядка 1 - α {quantil = }")
```

```python
p_value = dist_H0.sf(delta)

print(f"Получившееся значение p-value: {p_value}")
```

```python
_, axs = plt.subplots(1, 1, figsize=(10, 6))

axs.set_title("Мощность критерия")
axs.set_xlabel("x")
axs.set_ylabel("W(x)")

x = np.linspace(0, 5, 100)

axs.plot(x, dist_H0.sf(-x + quantil))

plt.grid(which='major', linewidth=0.8)

plt.show()
```
